```python
things = ["mozzarella", "cinderella", "salmonella"]
things[1] = things[1].capitalize()
print(things)
# 7.4 & 7.5; capitalizing it did change the element in the list
```

```python
things = ["mozzarella", "cinderella".capitalize(), "salmonella"]
things[0] = things[0].upper()
print(things)
# 7.6; making the cheesy element all uppercase
```

```python
things = ["mozzarella".upper(), "cinderella".capitalize(), "salmonella"]
del things[2]
print(things)
# 7.7; deleting the diseased element from the list
```
