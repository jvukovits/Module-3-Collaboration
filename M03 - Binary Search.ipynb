```python
class Solution:
    def binarysearch(self, arr, k):
        
        i = -1
        l, r = 0, len(arr) - 1
        

        while l <= r:
            mid = l + (r - l) // 2

            if arr[mid] == k:
                i = mid
                r = mid - 1
            elif arr[mid] < k:
                l = mid + 1
            else:
                r = mid - 1
                
        return i
```
