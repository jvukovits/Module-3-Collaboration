```python
class Solution:
    def sort012(arr):
        n = len(arr)

        low = 0 # boundary for the 0s
        mid = 0 # current element being checked
        hi = n - 1 # boundary for the 2s

        while mid <= hi:
            if arr[mid] == 0:
                # if mid is 0, swap with low, move both
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
                
            elif arr[mid] == 1:
                mid += 1
                # if mid is 1, it is in correct position
            else:
                arr[mid], arr[hi] = arr[hi], arr[mid]
                hi -= 1
                # if mid is 2, swap with hi, move hi back

    if __name__ == "__main__":
        arr = [0, 1, 2, 0, 1, 2]
        n = len(arr)

        sort012(arr)
        # sort the array

        for i in range(n):
            print(arr[i], end="")
```
